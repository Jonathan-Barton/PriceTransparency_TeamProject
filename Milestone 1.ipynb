* Our data has several elements. Each row contains information about the hospital it is linked to (# of beds, nonprofit/for-profit, yearly discharges, patient experience rating), as well as CPT codes and different prices. CPT codes are unique medical billing codes that are used to bill for specific procedures. For example, code 44950 is used to bill for an appendectomy. Each row has a gross price, which is typically the highest-paid price for a procedure, and most rows contain discounted cash prices, for customers who pay their bill in cash (cheapest option other than using insurance).
